```python
import pandas as pd

# Example: if you already have a df with these columns
# df = pd.DataFrame({'userId': [...], 'movieId': [...], 'rating': [...]})

# Step 1: Rescale ratings from [1, 5] â†’ [1, 10]
ratings['rescaled_rating'] = ratings['rating'] * 2

# Step 2: Group by userId and compute aggregates
user_stats = ratings.groupby('userId').agg(
    num_rated_movies=('movieId', 'count'),
    avg_rating_given=('rescaled_rating', 'mean')
).reset_index()

user_stats.head()
```

```python
import pandas as pd
import ast

# Load both datasets
movies = pd.read_csv("data/movies.csv")
new_movies = pd.read_csv("data/new_movies.csv")

# Convert genres column in new_movies to list of dicts if needed
def parse_genre_list(val):
    try:
        return ast.literal_eval(val) if isinstance(val, str) else val
    except:
        return None

new_movies['genres'] = new_movies['genres'].apply(parse_genre_list)

# Merge to get fallback title and genres from movies.csv
merged = new_movies.merge(movies[['movieId', 'title', 'genres']], on='movieId', how='left', suffixes=('', '_source'))

# Fill missing titles
merged['title'] = merged['title'].fillna(merged['title_source'])

# Convert genres string from source into list of dicts: [{'id': 0, 'name': 'Genre'}]
def convert_genres(row):
    val = row['genres']
    
    # If genres is None or NaN
    if val is None or (isinstance(val, float) and pd.isna(val)):
        fallback = row['genres_source']
        if pd.notna(fallback):
            genre_names = [g.strip() for g in fallback.split('|')]
            return [{'id': 0, 'name': name} for name in genre_names]
        else:
            return None

    # If genres is an empty list
    if isinstance(val, list) and len(val) == 0:
        fallback = row['genres_source']
        if pd.notna(fallback):
            genre_names = [g.strip() for g in fallback.split('|')]
            return [{'id': 0, 'name': name} for name in genre_names]
        else:
            return []

    return val


merged['genres'] = merged.apply(convert_genres, axis=1)

# Drop helper columns
merged = merged.drop(columns=['title_source', 'genres_source'])

# Final cleaned DataFrame
# merged.to_csv("cleaned_new_movies.csv", index=False)  # if needed
```
